**Imports**

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Data**

```python
# Load the data from your .csv.gz file into a new DataFrame.
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.info()
df_2001.head()
```

```python
# Load the data from your .csv.gz file into a new DataFrame.
df_2002 = pd.read_csv('Data/final_tmdb_data_2002.csv.gz')
df_2002.info()
df_2002.head()
```

# Merge 2001 and 2002 Data

```python
dfs = [df_2001, df_2002]
df_combined = pd.concat(dfs)
df_combined.info()
df_combined.head()
```

## How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
# find total movies with valid budget
filter_budget = df_combined['budget'] >0
filter_budget.sum()
```

```python
# find total movies with valid revenue
filter_revenue = df_combined['revenue'] >0
filter_revenue.sum()
```

```python
filter_financial = filter_revenue | filter_budget
filter_financial.sum()
```

```python
# Remove any movies with 0's for budget AND revenue from the remaining visualizations.
df_filtered = df_combined[filter_financial]
df_filtered.info()
df_filtered.head(2)
```

```python
print(f"There are {len(df_filtered)} movies with valid budget and revenue.")
```

## How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Display null matrix
import missingno as msno
msno.matrix(df_filtered);
```

```python
with pd.option_context('display.float_format',lambda x: f" {x:,.0f}"):
    display(df_filtered.groupby('certification')['imdb_id'].count())

sns.countplot(data=df_filtered, x='certification');
df_combined['certification'].value_counts(dropna=False);
```

## What is the average revenue per certification category?

```python
with pd.option_context('display.float_format',lambda x: f"$ {x:,.0f}"):
    display(df_filtered.groupby('certification')['revenue'].mean())

sns.barplot(data=df_filtered, x='certification', y='revenue');
```

## What is the average budget per certification category?

```python
with pd.option_context('display.float_format',lambda x: f"$ {x:,.0f}"):
    display(df_filtered.groupby('certification')['budget'].mean())

sns.barplot(data=df_filtered, x='certification', y='budget');
```
